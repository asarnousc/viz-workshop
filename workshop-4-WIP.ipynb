### 1. Avoid radial charts

Pie charts appear deceitfuly friendly, but in reality, the data can almost always be presented in a better format. The human eye has evolved to be able to compare linear distances. It is difficult to tell the difference between angles, especially when precision is important.

even for very few categories

```python
plotted_genres = movies.first_genre.value_counts().index[:8]
```

```python
split_year = 1990
genres_old = movies[movies.title_year <  split_year].first_genre.value_counts()[plotted_genres]
genres_new = movies[movies.title_year >= split_year].first_genre.value_counts()[plotted_genres]
```

```python
plt.subplot(1, 2, 1)
genres_old.plot('pie', title="Before '90s")

plt.subplot(1, 2, 2)
genres_new.plot('pie', title="After '90s")

plt.subplots_adjust(right=1.75)
```

which one is larger: 
 - before '90: crime or horror? 
 - after '90: comedy or action?
 - action: before '90s or after '90s?

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(12, 4))

(genres_old/sum(genres_old)).plot('bar', title="Before '90s", ax=ax1)
ax1.set_ylabel('% Movies')
ax1.grid(axis='y', linewidth=.5)

(genres_new/sum(genres_new)).plot('bar',title="After '90s", ax=ax2)
ax2.grid(axis='y', linewidth=.5)
```

### 2. Keep it simple

simple is most efficient at convetying the message

also 3D

also colors

ink to data ratio

do not sacrifice readability for "pretty"

### 3. Provide labels

axis, units, and a title, and a short explanation if possible (you should treat it as if a person randomly opens the boook and lands on your chart: they should be able to understand what's going on), and add an accompanying conclusion

### 4. Axes ranges

from zero: but not always

Bar charts ask viewers compare the relative lengths. Showing just the tips of the bars to exaggerate differences in the data, their relative lengths changes. So people are either misled and take away the wrong message, or end up having to read the numbers, which defeats the purpose of the chart.

```python
sw_movies = movies[movies.movie_title.str.contains('Star Wars')].dropna().set_index('movie_title')
```

### 5. 

Line charts show a trend. Stretching the height of the graph can overblow the message, while stretching the width can understate it. This leads to misleading the audience.

6.

It's harder to compare areas than lengths. We usually underestimate the size of bigger shapes and overestimate the size of the smaller ones, because we instinctively judge the lengths or widths and not their areas.

units:
 - too many decimal places
 - 0.5 instead of 50%
 - too many zeros instead of "k" or "m", etc

### Other

- plot at most 8 categorical colors
- investigate spurious correlations=pvalue (and check if your model is complex enough=line fitting both semi circle and line)
- never use non-linear interpolation: creates points that aren't there
- pay attention to the kind of average (example with twin peaks distribution and outlier):
 - mean
 - median
 - mode
- ignore irrelevant outliers:
 - stds from mean
 - percentile cutoff
 - iqr
 - replace with min/max value
